```python
import re

text: str = "Hola mundo!. Esto es una prueba, pero en español."
list_text_elements: list[str] = re.split(pattern=r'(\s|[\\^$.,|?!*+()\[\]{}])', string=text)
list_text_elements
```

```python
# Palabras (letras/números) o símbolos específicos
# \w+ captura palabras completas.
# [\\^$.,|?!*+()\[\]{}] captura los símbolos uno a uno.
pattern = r"[^\W_]+|[.,|?!*+\[\]{}'\"():;]+"
content_file_cleaned = re.findall(pattern, content_file, re.UNICODE)
content_file_cleaned
```

```python
vocab_dictionary: dict = {content: index for index, content in enumerate(all_worlds)}
vocab_dictionary
```

```python
class SimpleTokenizerV1:

    def __init__(self, vocab_dictionary: dict) -> None:

        self._pattern = r"[^\W_]+|[.,|?!*+\[\]{}'\"():;]+"

        self.vocab_dictionary = vocab_dictionary
        self.int_to_str = {index: content for content, index in vocab_dictionary.items()}

    def encode(self, text_content: str) -> list[int]:

        content_file_cleaned: list[str] = re.findall(
            pattern=self._pattern, 
            string=text_content,
            flags=re.UNICODE
        )

        return [self.vocab_dictionary[elem] for elem in content_file_cleaned]

    def decode(self, id_content: str) -> list[str]:

        text = " ".join([self.int_to_str[_id] for _id in id_content])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
decoded_text: list[str] = tokenizer.decode(id_content=list_ids)
decoded_text
```

# Tokenizer with unknown tokens

```python
class SimpleTokenizerV2:

    def __init__(self, vocab_dictionary: dict) -> None:

        self._pattern = r"[^\W_]+|[.,|?!*+\[\]{}'\"():;]+"

        self._special_tokens: list[str] = sorted(
            {
                "<|unk|>",  
                "<|endoftext|>" 
            }
        )

        self.vocab_dictionary = self._add_special_tokens(
            vocab_dictionary=vocab_dictionary
        )

        self.int_to_str = {index: content for content, index in self.vocab_dictionary.items()}

    def _add_special_tokens(self, vocab_dictionary: dict) -> dict:

        copy_vocab_dictionary = vocab_dictionary.copy()

        for token in self._special_tokens:
            copy_vocab_dictionary.update({token: len(copy_vocab_dictionary)})

        return copy_vocab_dictionary

    def encode(self, text_content: str) -> list[int]:

        content_file_cleaned: list[str] = re.findall(
            pattern=self._pattern, 
            string=text_content,
            flags=re.UNICODE
        )

        return [self.vocab_dictionary.get(elem, self.vocab_dictionary["<|unk|>"]) for elem in content_file_cleaned] + [self.vocab_dictionary["<|endoftext|>"]]

    def decode(self, id_content: str) -> list[str]:

        text = " ".join([self.int_to_str[_id] for _id in id_content])
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

# Data loader

```python
import torch
from torch.utils.data import Dataset, DataLoader
from tiktoken.core import Encoding
```

```python
class GPTDatasetV1(Dataset):

    def __init__(self, text: str, tokenizer: Encoding, max_length: int, stride: int) -> None:
        """_summary_

        Args:
            text (str): Texto a tokenizar.
            tokenizer (Encoding): Tokenizador.
            max_length (int): Longitud máxima de cada fila del dataset (N X M),
                N es el numero de input, M la longitud del input, pues max_length es M.
            stride (int): Tamaño de paso entre secuencias de IDs.
        """

        self.input_ids: list[int] = []
        self.target_ids: list[int] = []

        list_tokens: list[int] = tokenizer.encode(text=text)

        for index in range(0, len(list_tokens) - max_length, stride):
            input_chunk: list[int] = list_tokens[index:index + max_length]
            target_chunk: list[int] = list_tokens[index + 1:index + 1 + max_length]
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self) -> int:
        return len(self.input_ids)

    def __getitem__(self, index: int) -> tuple[int, ...]:
        return self.input_ids[index], self.target_ids[index]
```

```python
def create_data_loader_v1(
    text: str,
    batch_size: int = 4,
    max_lenght: int = 256,
    stride: int = 128,
    shuffle: bool = True,
    drop_last: bool = True,
    num_workers: int = 0
) -> DataLoader:

    tokenizer = tiktoken.get_encoding("gpt2")
    
    dataset = GPTDatasetV1(
        text=text,
        tokenizer=tokenizer,
        max_length=max_lenght,
        stride=stride
    )

    return DataLoader(
        dataset=dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=drop_last,
        num_workers=num_workers
    )
```
